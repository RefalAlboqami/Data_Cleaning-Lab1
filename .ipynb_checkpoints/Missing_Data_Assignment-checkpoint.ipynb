### 1 dropping Columns ( Employee_Name, DOB, DateofTermination)

```python
df.drop(columns=['Employee_Name'], inplace=True)
df.drop(columns=['DOB'], inplace=True)
df.drop(columns=['DateofTermination'], inplace=True)

df.info()
```

### 2 Filling by Mean for Columns (Salary,Absences)

```python
mean_value1=df['Salary'].mean() 
df['Salary'].fillna(value=mean_value1, inplace=True)

mean_value2=df['Absences'].mean() 
df['Absences'].fillna(value=mean_value2, inplace=True)
df.info()
```

### 3 Replace the NaN with another values for (Sex,ManagerID)

```python
df['ManagerID'].fillna(0, inplace=True)
df['Sex'].fillna('Undefined', inplace=True)

df.info()
```
